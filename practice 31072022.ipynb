```python
y=df_new['Exited']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# training and testing the data:
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm=LogisticRegression()
lm.fit(x_train,y_train)
```

```python
y_predit=lm.predict(x_test)
```

# Ways to Evaluate Regression Models :

# Accuracy_score:

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_predit))
```

# R2 Score:

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_predit))
```

# Mean squared error:

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test,y_predit))
```

# Mean Absolute error:

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,y_predit))
```

```python
df=pd.DataFrame(data={"Actual value":y_test,"Predict value":y_predit})
df["Difference"]=df["Predict value"]-df["Actual value"]
```

```python
df
```

```python
squared=np.square(df["Difference"])
squared.mean()
```

# How to Evaluate a Classification Model:

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predit))
```
